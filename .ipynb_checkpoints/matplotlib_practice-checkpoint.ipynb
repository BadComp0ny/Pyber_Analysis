```python
%matplotlib inline
```

```python
# import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot using Object-oriented interface method "ax.plot()"
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot using OoI method in an alternative way
fig = plt.figure()
ax = fig.add_subplot()
```

```python
# Create the plot using OoI method in an alternative way
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Way to create a quick graph usings OoI method with ax.plot()
ax = plt.axes()
ax.plot(x_axis, y_axis)

#To change figure-level attributes (such as axis labels, a title,
#or a legend) or save the figure as an image, use fig = plt.figure().
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()

# plt.show() should only be used once per session because it can cause the graph to display improperly. 
# If you have problems with data not being displayed properly, you may have to restart your kernel and clear your output.
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker = "*", color = "blue", linewidth = "2", label='Boston')

# Create labels for the x & y axis
plt.xlabel('Date')
plt.ylabel('Fare($)')

# Set the y limit between 0 & 45
plt.ylim(0, 45)

# Create a title
plt.title("Pyber Fare by Month")

# Add a grid
plt.grid()

# Add the legend
plt.legend()
```

```python
# Create the plot and add a label for the legend
ax = plt.axes()
ax.plot(x_axis, y_axis, color="green", marker="d", linewidth="2", label = "Boston")

# Create labels for the axes
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")

# Create a Title
ax.set_title("Pyber Fare by Month")

# Create y-axis limit
ax.set_ylim(0,45)

# add grid lines
ax.grid()

# add a legend
ax.legend()
```

```python
# Create verticle bar chart using ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis,y_axis)
```

```python
# Create a horizontal chart using ax.plt()
fig,ax = plt.subplots()
ax.barh(y_axis,x_axis)
```

```python
# Create a horizontal chart using ax.plt()
fig,ax = plt.subplots()
ax.barh(x_axis, y_axis, color = "cyan", label="Chicago")
ax.invert_yaxis()

# add labels to axis
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")

# add title
ax.set_title("Pyber Fare by Month")

# add legend
ax.legend()
```

```python
# Create a scatter plot using MATLAB method.
plt.plot(x_axis,y_axis, 'o')
```

```python
# Create a scatter plot using plt.scatter()
plt.scatter(x_axis, y_axis)
```

```python
# Create a scatter plot using MATLAB method.
plt.plot(y_axis,x_axis, 'o', color="red", label="Chicago")
plt.gca().invert_yaxis()

# Label the title and the axes
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("Pyber Fare by Month")

# set limit for x axis
plt.xlim(0,50)

plt.legend()
```

```python
# Create a bubble chart using MATLAB method
plt.scatter(x_axis,y_axis, s=y_axis)
```

```python
# iterate through y axis data and multiply each point by 3
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
    
# Create new bubble chart
plt.scatter(x_axis,y_axis, s=y_axis_larger)
```

```python
#Use list comprehension method to repeat above chart
plt.scatter(x_axis,y_axis, s = [i *3 for i in y_axis])
```

```python
# Create a scatter plot using the OoI ax.plot() method.
fig, ax = plt.subplots()
ax.scatter(x_axis,y_axis)
```

```python
# Create a bubble plot using the OoI ax.plot() method
fig, ax = plt.subplots()
ax.scatter(x_axis,y_axis, s=y_axis)
```

```python
# Create a bubble plot using the OoI ax.plot() method
fig, ax = plt.subplots()
ax.scatter(y_axis,x_axis, s=[i * 5 for i in y_axis], c="blue", edgecolors="k", linewidths=2, alpha=.20, label="Boston")
ax.legend()

ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.set_title("Pyber Fare by Month")
ax.set_xlim(0,50)
ax.invert_yaxis()
```

```python
# Create a Pie Chart using the MATLAB method
plt.pie(y_axis, labels=x_axis)
plt.show
```

```python
explode_values = ( 0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ['slateblue', 'magenta', 'lightblue', 'green', 'yellowgreen', 'greenyellow', 'yellow', 'orange', 'gold', 'indianred', 'tomato', 'mistyrose']
explode_values = ( 0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis, explode=explode_values, colors=colors, labels=x_axis, autopct='%.1f%%')

plt.show()
```

```python
# Create a Pie Chart using ax.plot()
fig, ax = plt.subplots()
ax.pie(y_axis, labels=x_axis)


plt.show()
```

```python
# Create a Pie Chart using ax.plot()
fig, ax = plt.subplots(figsize=(8,8))
explode_values = (0,0,0.2,0,0,0,0.2,0,0,0,0,0)
colors = ['slateblue', 'magenta', 'lightblue', 'green', 'yellowgreen', 'greenyellow', 'yellow', 'orange', 'gold', 'indianred', 'tomato', 'mistyrose']
ax.pie(y_axis, labels=x_axis, colors=colors, explode=explode_values, startangle=-90,shadow=True, autopct='%.1f%%')
ax.invert_yaxis()

plt.show()
```
